```python
import arcpy

# Set the path to the layer
layer = "OrganicsData_V1"

# Start an update cursor to calculate the field
with arcpy.da.UpdateCursor(layer, ["lyr_14C", "lyr_14C_01"]) as cursor:
    for row in cursor:
        # Check the value of the 'lyr_14C' field and set 'lyr14C_01' accordingly
        if row[0] == "yes":
            row[1] = 1
        elif row[0] == "no":
            row[1] = 0
        cursor.updateRow(row)

print("Field 'lyr14C_01' updated successfully.")
```
